```python
import os
import csv

def read_kotlin_files(project_dir, drive_csv_path):
    # Open CSV file in append mode
    with open(drive_csv_path, mode='a', newline='', encoding='utf-8') as csv_file:
        writer = csv.writer(csv_file)

        # Check if the file is empty, write header only if file is new
        if os.stat(drive_csv_path).st_size == 0:
            writer.writerow(['FILE_PATH', 'CONTENT'])  # Write the header row

        # Walk through the directory recursively
        for root, dirs, files in os.walk(project_dir):
            for file in files:
                if file.endswith('.kt'):
                    file_path = os.path.join(root, file)

                    # Read the file content
                    with open(file_path, 'r', encoding='utf-8') as f:
                        content = f.read()

                    # Write the file path and content to the CSV file
                    writer.writerow([file_path, content])

    print(f"File updated on Drive at: {drive_csv_path}")

# Define the project directory and the CSV file path directly on Google Drive
project_directory = '/content/WeCal'  # Update this path accordingly
drive_csv_file = '/content/drive/MyDrive/evaluate.csv'

# Run the function to append content to the CSV stored in Google Drive
read_kotlin_files(project_directory, drive_csv_file)


def count_entries(drive_csv_file):
    # Open the CSV file directly from Google Drive and count the number of entries
    with open(drive_csv_file, mode='r', encoding='utf-8') as file:
        reader = csv.reader(file)
        next(reader)  # Skip the header row
        entry_count = sum(1 for row in reader)  # Count the rows

    return entry_count

# Get the number of entries
num_entries = count_entries(drive_csv_file)
print(f'Number of entries (excluding header): {num_entries}')
```

```python
import csv
import re

def remove_comments(content):
    # Remove single-line comments (//)
    content = re.sub(r'//.*', '', content)

    # Remove multi-line comments (/* */)
    content = re.sub(r'/\*.*?\*/', '', content, flags=re.DOTALL)

    return content

def count_non_empty_lines_without_comments(csv_file):
    total_non_empty_lines = 0

    # Open the CSV file and read its contents
    with open(csv_file, mode='r', encoding='utf-8') as file:
        reader = csv.reader(file)
        next(reader)  # Skip the header row

        # Iterate over each row in the CSV file
        for row in reader:
            file_path, content = row

            # Remove comments from the content
            content_without_comments = remove_comments(content)

            # Split the content into lines
            lines = content_without_comments.split('\n')

            # Count non-empty lines (ignoring lines that contain only whitespace)
            non_empty_lines = sum(1 for line in lines if line.strip())

            # Add to the total count
            total_non_empty_lines += non_empty_lines

    return total_non_empty_lines

# Define the path to the CSV file
drive_csv_file = '/content/drive/MyDrive/evaluate.csv'

# Get the number of non-empty lines excluding comments
num_non_empty_lines_without_comments = count_non_empty_lines_without_comments(drive_csv_file)
print(f'Total number of non-empty lines (excluding comments): {num_non_empty_lines_without_comments}')
```
